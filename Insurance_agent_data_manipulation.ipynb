# Task #8
Mobile & Travel Insurance premium are discounted by 10%
Health and Property Insurance premium are increased by 7%
Life and Motor Insurance premium are marginally increased by 2%
What will be overall change in % of the Premium Amount Collected for all these Customer?


```python
initial_premium_amount_sum = insurance_data['PREMIUM_AMOUNT'].sum()
initial_premium_amount_sum
```

```python
mobile_travel = insurance_data[(insurance_data.INSURANCE_TYPE == 'Travel') | (insurance_data.INSURANCE_TYPE == 'Mobile')]
```

```python
mobile_travel_discounted = mobile_travel.PREMIUM_AMOUNT.sum() - (mobile_travel.PREMIUM_AMOUNT.sum()* 10/100)
```

```python
health_property = insurance_data[(insurance_data.INSURANCE_TYPE == 'Health') | (insurance_data.INSURANCE_TYPE == 'Property')]
```

```python
health_property_discounted = health_property.PREMIUM_AMOUNT.sum() + (health_property.PREMIUM_AMOUNT.sum()* 7/100)
```

```python
life_motor = insurance_data[(insurance_data.INSURANCE_TYPE == 'Life') | (insurance_data.INSURANCE_TYPE == 'Motor')]
```

```python
life_motor_discounted = life_motor.PREMIUM_AMOUNT.sum() + (life_motor.PREMIUM_AMOUNT.sum()* 2/100)
```

```python
Final_Total_premium_amount = mobile_travel_discounted+health_property_discounted+life_motor_discounted
Final_Total_premium_amount
```

```python
Total_percentage_change =(( Final_Total_premium_amount - initial_premium_amount_sum)/initial_premium_amount_sum)*100
```

```python
print(Total_percentage_change)
```

# Task #9

19. They have laid down an eligibility Criteria as follow
IF CUSTOMER_TENURE > 60 AND EMPLOYMENT_STATUS = “N”

Create a new column “ELIGIBLE_FOR_DISCOUNT” and find it

# Task #12

Business wants to find all Suspicious Employees (Agents).
IF TOTAL CLAIM AMOUNT which meet below criteria is >= 15000 THEN AGENT is classified
as Suspicious ELSE Not
CLAIM_STATUS = Approved AND CUSTOMER_RISK_SEGMENTATION = High
AND INCIDENT_SEVERITY = “Major Loss”
If Suspicious, then 1 ELSE 0. Find mean of this column.

```python
insurance_data['suspicios'] = insurance_data.apply(lambda row: 1 if row['CLAIM_AMOUNT'] >= 15000 and row['CLAIM_STATUS'] == "A" and row['RISK_SEGMENTATION'] == 'H' and row['INCIDENT_SEVERITY'] == 'Major Loss' else 0 ,axis =1) 
```
